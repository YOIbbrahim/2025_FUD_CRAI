#EXPLORATORY DATA ANALYSIS

```python
# plotting barchart
sns.countplot(x='species', data=df)
plt.title("Bar Chart")
plt.show()
```

```python
# plotting pie chart
plt.pie(df["species"].value_counts(), labels = df["species"].unique(), autopct="%1.1f%%")
plt.title("piechart")
plt.show()
```

```python
#Scatterplot
sns.scatterplot(x='sepal_width', y='sepal_length', hue="species", data=df)
plt.title('scatterplot for the three species')
plt.show()
```

```python
# plotting histogram
sns.histplot(df, kde=True)
plt.title("histogram")
plt.show()
```

```python
# line chart
sns.lineplot(data=df)
plt.title("line chart")
plt.show()
```

```python
df1=df.drop("species", axis=1)
```

```python
# heat map
sns.heatmap(df1.corr(), annot=True)
plt.title("heatmap")
plt.show()
```
